# F-test

If the energy on the transverse component is drawn from a chi-square distribution, then the ratio R is drawn from an $F_{k,\nu-k}$ distribution:

$R(\phi, \delta t) = \frac{(\lambda_2 (\phi, \delta t) - \lambda_{2}^{min}) / k}{\lambda_{2}^{min}/(\nu-k)} \sim F_{k,\nu-k}$

$k=2$ the number of estimated parameters ($\phi, \delta t$), and $\nu$ is the number of degrees of freedom.

```python
import splitwavepy as sw
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
alpha = np.linspace(0,1,100)
k = 2
ndf = 100
for ndf in range(5,30,5):
    F = stats.f.ppf(alpha,k,ndf)
    plt.plot(alpha, F, label=str(ndf), alpha=0.7)
    plt.axhline(stats.f.ppf(0.95,k,ndf), linewidth=0.5, color='k')

plt.title('F point percentile function')
plt.xlabel('Quantile')
plt.ylabel('R')
plt.legend()

plt.axvline(0.95, linewidth=0.5, color='k')
plt.show()
```

```python
x = np.linspace(0,10,100)
k = 2
for ndf in range(5,25,5):
    F = stats.f.cdf(x,k,ndf)
    plt.plot(x, F, label=str(ndf), alpha=0.7)

plt.title('F cumulative density function')
plt.xlabel('R')
plt.ylabel('Quantile')
plt.legend()
plt.show()
```

```python
x = np.linspace(1,10,100)
k = 2
for ndf in range(5,25,5):
    F = stats.f.pdf(x,k,ndf)
    plt.plot(x, F, label=str(ndf), alpha=0.7)

plt.title('F probability density function')
plt.xlabel('R')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
